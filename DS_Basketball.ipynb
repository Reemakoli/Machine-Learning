<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">3. Data preprocessing :</span>

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">3.1 Dealing with Duplicates data</span>

```python
##Identify duplicates records in the data
dupes = df_Teams_Basketball.duplicated()
sum(dupes)
```

```python
#Column names except "Team" as it is categorical
cols = df_Teams_Basketball[df_Teams_Basketball.columns.difference(['Team'])].columns
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">3.2 Replace '-' in columns with 0</span>

```python
#Replace '-' in columns with 0 except Column "Teams"

df_Teams_Basketball.loc[:,cols]= df_Teams_Basketball.loc[:,cols].replace('-', '0', regex=True)
df_Teams_Basketball
```

```python
#Convert date ranges to their “start date”, wherever present: 1931to32; 1956-57, 38-54
#Convert the column Object datatype to int

df_Teams_Basketball.loc[:,'TeamLaunch']  = df_Teams_Basketball['TeamLaunch'].str.extract(r'^(\d{4})', expand=False)
df_Teams_Basketball
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">3.3 Convert required columns to numeric</span>

```python
#Convert all columns to numeric except "Team"
df_Teams_Basketball[cols] = df_Teams_Basketball[cols].apply(pd.to_numeric, errors='coerce', axis=1)
df_Teams_Basketball.info()
```

```python
# select numeric columns
df_numeric = df_Teams_Basketball.select_dtypes(include=[np.number])
numeric_cols = df_numeric.columns.values
print(numeric_cols)
```

```python
# select non numeric columns
df_non_numeric = df_Teams_Basketball.select_dtypes(exclude=[np.number])
non_numeric_cols = df_non_numeric.columns.values
print(non_numeric_cols)
```

```python
# drop the rows which have null data          
df_Teams_Basketball = df_Teams_Basketball.dropna()
```

```python
#deleting duplicate rows
df_Teams_Basketball = df_Teams_Basketball.drop_duplicates()
```
