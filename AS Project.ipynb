<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;"> 2. Question: An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions.

A. Probability that none of the items are defective?

B. Probability that exactly one of the items is defective?

C. Probability that two or fewer of the items are defective?

D. Probability that three or more of the items are defective ?
</span>

```python
# setting the values
n  = 10 # a random sample of 10 manufactured items is selected 

p  = 5/100 # the failure rate for the manufactured item is 5%

k =np.arange(0,10) # # An array of quantiles
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;"> Binomial Distribution</span>

```python
#Binomial Distribution
binomialDistibution= stats.binom.pmf(k,n,p)
print(binomialDistibution)
```

```python
plt.plot(k,binomialDistibution,'o-')
plt.title('Binomial Distribution: n=%i , p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Number of Defective Items')
plt.ylabel('Probability of Defective Items')
plt.show()
```

```python
print('A. Probability that none of the items are defective is %1.4f'%binomialDistibution[0])
```

```python
print('B. Probability that exactly one of the items is defective is %1.4f'%binomialDistibution[1])
```

```python
#Cumulative DistributionFunction
cumulativeDistibution= stats.binom.cdf(k,n,p)
print(cumulativeDistibution)
```

```python
print('C. Probability that two or fewer of the items are defective is %1.4f'%cumulativeDistibution[2])
```

```python
defectiveItems = 1-cumulativeDistibution[2]
print('D. Probability that three or more of the items are defective is %1.2f'%defectiveItems)
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">3. Question: A car salesman sells on an average 3 cars per week.

A. Probability that in a given week he will sell some cars.

B. Probability that in a given week he will sell 2 or more but less than 5 cars.

C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold perweek.
</span>

```python
#average rate, denoted with lambda in the formula for poisson distribution 
rate = 3

#a numpy array of different no. of successes ranging from 0 to 19 to construct a probability distribution
n = np.arange(0,16)

#calculating the probability distribution for different values of x
poisson = stats.poisson.cdf(n,rate)

print(poisson)
```

```python
# P = probability that the car salesman sells car
sellingProbability  = 1 - poisson[0]
print('A. Probability that in a given week he will sell some cars is %1.4f' % sellingProbability)
```

```python
# Let P1 be the probability that the car salesman sells  more than 4 cars
P1 = poisson[4] # P(X >=4)
# Let P2 be the probability that the car salesman sellss more than 1 car
P2 = poisson[1] # P(X >=1)
P = P1 - P2 # Prob. that the car salesman sells between 2 and 5 car
print('B. Probability that in a given week he will sell 2 or more but less than 5 cars is %1.4f' % P)
```

 C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold perweek.

```python
#visualising the plot with pyplot

plt.plot(n,poisson,'o-')
plt.title('Poisson: $\lambda$ = %i ' % rate)
plt.xlabel('Number of Cars sold per week')
plt.ylabel('Cumulative Probability of cars sold per week')
plt.show()
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;"> 5. Question: A group of 300 professionals sat for a competitive exam. The results show the information of marks obtained by them have a mean of 60 and a standard deviation of 12. The pattern of marks follows a normal distribution. Answer the following questions

A. What is the percentage of students who score more than 80.

B. What is the percentage of students who score less than 50.

C. What should be the distinction mark if the highest 10% of students are to be awarded distinction?
</span>

```python
n = 500
mean = 60
stdDev = 12
```

```python
#Calculate Normal Distribution

data=np.arange(10,110,0.1)
pdf= stats.norm.pdf(x = data,loc = mean,scale = stdDev)
sns.set_style('whitegrid')
sns.lineplot(data, pdf , color = 'black')
plt.xlabel('Marks obtained')
plt.ylabel('Probability Density')
plt.show()
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;"> A. What is the percentage of students who score more than 80
 </span>


```python
cdf_value=stats.norm.cdf(x=80,loc=mean,scale=stdDev)
z_score_1 = 1- cdf_value

print(f'The percentage of students who score more than 80 is {z_score_1:1.4} or {z_score_1 * 100:.2f} %' )
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;"> B. What is the percentage of students who score less than 50.</span>

```python
z_score_2=stats.norm.cdf(x=50,loc=mean,scale=stdDev)

print(f'The percentage of students who score less than 50 is {z_score_2:1.4} or {z_score_2 * 100:.2f}%')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;"> C. What should be the distinction mark if the highest 10% of students are to be awarded distinction?</span>

```python
#convert the percentage to a z-score
z_score=stats.norm.ppf(0.90)

#z_score=(x-mean)/stdDev
x = (z_score * stdDev) + mean

print(f'The distinction mark if the highest 10% of students are to be awarded distinction is {x:1.4}')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">4. Statistical analysis</span>

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">4.1 Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?
Write the null hypothesis and alternative hypothesis.
Test for significance and conclusion</span>

- Visually the distributions are quiet similar but are they statistical significant 
-There are two ways we can check if the difference is significant
    - z-test 
    - t-test
- The distributions are not normal yet, since our sample size is reasonably big ( >30 samples)
- To prove : Is there significant difference between Funds raised by companies that are still operating vs companies that closed down

**Null hypothesis (Ho)** : There is no difference between the two means of Funds raised by companies

**Alternate hypothesis (Ha)** : There is significant difference between the two means of Funds raised by companies

```python
sample_Operating = df_startups.loc[df_startups.OperatingState == 'Operating', 'Funds_in_million']
sample_Closed = df_startups.loc[df_startups.OperatingState == 'Closed', 'Funds_in_million']

alpha = 0.05   # Let's consider a significance level of 5%
```

```python
from statsmodels.stats.weightstats import ztest

stats, p_value = ztest(sample_Operating, sample_Closed)

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
    
```

```python
from scipy.stats import ttest_ind

stats, p_value = ttest_ind(sample_Operating, sample_Closed)

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

**- From the tests above we found there is no statisticcal significance difference  between  Funds raised by companies that are still operating vs companies that closed down i.e no evidence to state that the companies that have raised more money tend to succeed more or vice-versa**

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">4.4 Calculate percentage of winner that are still operating and percentage of contestants that are still operating</span>

```python
winners = df_startups.Result.value_counts()[1:].sum()
contestants = df_startups.Result.value_counts()['Contestant']

contestants_operating = df_startups.OperatingState[df_startups.Result == 'Contestant'].value_counts().loc['Operating']
winners_operating = df_startups.OperatingState[df_startups.Result != 'Contestant'].value_counts().loc['Operating']

print(f'The percentage of winner that are still operating is {(winners_operating/winners)*100: 1.4} %')
print(f'The percentage of contestants that are still operating is {(contestants_operating/contestants)*100: 1.4} %')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">4.5 Write your hypothesis comparing the proportion of companies that are operating between winners and contestants:
Write the null hypothesis and alternative hypothesis.
Test for significance and conclusion</span>

- The Proportion of winner that are still operating is  76.44 %
- The Proportion of contestants that are still operating is  68.03 %
- The proportions are different but are they statistically significant?
- There are two ways we can check if the difference is significant
    - Z-test for proportions
    - Chi-square test

**Null hyputhesis (Ho)**: The proportion of companies that are operating is the same in both categories - winners and contestants

**Alternative hypothesis (Ha)**: The proportion of companies that are operating is significantly different from each other, among the two categories

```python
#z- test
from statsmodels.stats.proportion import proportions_ztest

stats, p_value = proportions_ztest([contestants_operating, winners_operating], [contestants, winners])

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

```python
#chi square test
#from scipy.stats import chi2_contingency
import statsmodels.stats.proportion as  proportion

stats, p_value, table = proportion.proportions_chisquare([contestants_operating, winners_operating], [contestants, winners])
if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">4.7 Write and perform your hypothesis along with significance test comparing the funds raised by companies across NY, SF and EU events from
2013 onwards.</span>

```python
#variables hold the amount of funds with respect to the Cities where the event was organised
NY_events = df_startups.loc[events[events.apply(lambda x: 'NY' in x)].index, 'Funds_in_million']
SF_events = df_startups.loc[events[events.apply(lambda x: 'SF' in x)].index, 'Funds_in_million']
EU_events = df_startups.loc[events[events.apply(lambda x: 'EU' in x or 'London' in x)].index, 'Funds_in_million']

print(f'Sample Size of Events in City NY :{len(NY_events)},SF: {len(SF_events)},EU: {len(EU_events)}')
```

- The sample sizes are different across the three categories of cities and it is okay to perform ANOVA on samples of different sizes


**Null Hypothesis(Ho)**: Average funds raised by companies across three cities are the same 

**Alternative Hypothesis(Ha)**: Average funds raised by companies across three cities are the different

```python
#ANOVA Test
from scipy.stats import f_oneway

stat, p_value = f_oneway(NY_events, SF_events, EU_events)

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

- The distribution of funds raised by the companies across the three regions is the same
- We found no evidence to say companies participating in certain regions have funds either significantly on the higher side or on the lower side
